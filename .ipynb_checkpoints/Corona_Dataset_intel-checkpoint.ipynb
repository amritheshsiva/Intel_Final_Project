```python
from __future__ import print_function
import os
data_path = ['data']
```

```python
import pandas as pd

# Import the data using the file path
filepath = os.sep.join(data_path + ['Covid Data.csv'])
data = pd.read_csv('Covid Data.csv')
data
```

```python
data.head(1).T
```

```python
data.drop(['DATE_DIED','CARDIOVASCULAR','OBESITY','RENAL_CHRONIC','TOBACCO','CLASIFFICATION_FINAL','ICU'], axis=1, inplace=True)
data
```

```python
data.drop(['OTHER_DISEASE','HIPERTENSION'], axis=1, inplace=True)
data
```

```python
data.fillna(0, inplace=True)
data
```

```python
data.drop(['INMSUPR'], axis=1, inplace=True)
data
```

```python
data.fillna(0, inplace=True)
data
```

```python
x_cols = [x for x in data.columns if x != "ASTHMA"]

X_data = data[x_cols]
y_data = data["ASTHMA"]
X_data
y_data
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)

knn = knn.fit(X_data, y_data)

y_pred = knn.predict(X_data)
```

```python
def accuracy(real, predict):
    return sum(y_data == y_pred) / float(real.shape[0])
```

```python
print(accuracy(y_data, y_pred))
```
